```python
from ultralytics import YOLO
import cv2

# Load panel detector (download from HF or Roboflow)
panel_model = YOLO("best.pt")  # or local .pt path

img = cv2.imread("003.jpg")
results = panel_model(img, conf=0.3)

# Draw and save
annotated = results[0].plot()
cv2.imwrite("panels_detected2.jpg", annotated)
print(f"Found {len(results[0].boxes)} panels")
```
