```python
from ultralytics import YOLO
from manga_ocr import MangaOcr
import cv2
from PIL import Image
```

```python
bubble_detector = YOLO("models/comic-speech-bubble-detector.pt")
```

```python
image_file = "img/005.jpg"
```

```python
img = cv2.imread(image_file)
```

```python
results = bubble_detector(img, conf=0.3)
```

```python
print("\nReading text from bubbles:\n")
bubble_number = 1
```

```python
ocr = MangaOcr()
```

```python
for box in results[0].boxes:
    # Get coordinates
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    
    # Crop the bubble
    bubble_crop = img[y1:y2, x1:x2]
    
    #change to pil format
    bubble_pil = Image.fromarray(bubble_crop)
    
    # Read Japanese text
    text = ocr(bubble_pil)
    print(f"Bubble {bubble_number}:")
    print(text)
    print("─" * 40)
    bubble_number += 1

print(f"\nFinished! Found and read {bubble_number-1} bubbles.")
```
